# Task 1

The team at RealAgents knows that the city that a property is located in makes a difference to the sale price. 

Unfortuntately they believe that this isn't always recorded in the data. 

Calculate the number of missing values of the `city`. 

 - You should use the data in the file "house_sales.csv". 

 - Your output should be an object `missing_city`, that contains the number of missing values in this column. 

```python
# Use this cell to write your code for Task 1
import pandas as pd
pd.set_option('display.max_columns', None)
clean_data = pd.read_csv('house_sales.csv')
#exploring the data
print(clean_data.head())
print(clean_data.info())
#looking for null values in the columns
print(clean_data.isna().sum())
#examining the city column
print(clean_data['city'].value_counts())
#the missing values in the city have --. we will count those values
missing_city = len(clean_data[clean_data['city'] == '--'])
print(missing_city)
```

# Task 2 

Before you fit any models, you will need to make sure the data is clean. 

The table below shows what the data should look like. 

Create a cleaned version of the dataframe. 

 - You should start with the data in the file "house_sales.csv". 

 - Your output should be a dataframe named `clean_data`. 

 - All column names and values should match the table below.


| Column Name | Criteria                                                |
|-------------|---------------------------------------------------------|
| house_id    | Nominal. </br> Unique identifier for houses. </br>Missing values not possible. |
| city        | Nominal. </br>The city in which the house is located. One of 'Silvertown', 'Riverford', 'Teasdale' and 'Poppleton' </br>Replace missing values with "Unknown". |
| sale_price  | Discrete. </br>The sale price of the house in whole dollars. Values can be any positive number greater than or equal to zero.</br>Remove missing entries. |
| sale_date   | Discrete. </br>The date of the last sale of the house. </br>Replace missing values with 2023-01-01. |
| months_listed  | Continuous. </br>The number of months the house was listed on the market prior to its last sale, rounded to one decimal place. </br>Replace missing values with mean number of months listed, to one decimal place. |
| bedrooms    | Discrete. </br>The number of bedrooms in the house. Any positive values greater than or equal to zero. </br>Replace missing values with the mean number of bedrooms, rounded to the nearest integer. |
| house_type   | Ordinal. </br>One of "Terraced", "Semi-detached", or "Detached". </br>Replace missing values with the most common house type. |
| area      | Continuous. </br>The area of the house in square meters, rounded to one decimal place. </br>Replace missing values with the mean, to one decimal place. |

```python
#CITY COLUMN CLEANING
#checking the datatype of the city column
print(f"The data type for the city column is : {clean_data['city'].dtype}")
#replacing incorrect city values with Unknown
clean_data['city'] = clean_data['city'].str.replace('--','Unknown')
#examining changes
print(clean_data['city'].value_counts())

#SALE_PRICE COLUMN CLEANING
#checking the datatype of the sale_price column
print(f"The data type for sale_price column is : {clean_data['sale_price'].dtype}")
#checking for values less than or equal to zero in the sale_price column
incorrect_sale_price = clean_data[clean_data['sale_price'] <= 0]
print(f"The number of incorrect sale_price values are: {len(incorrect_sale_price)}")

#SALE DATE COLUMN CLEANING
#checking the datatype of the sale_date column
print(f"The data type for the sale_date column is : {clean_data['sale_date'].dtype}")
#checking for missing values in the  sale_date column
missing_sale_date = clean_data['sale_date'].isna().sum()
print(f"The number of missing sale_date values is: {missing_sale_date}")

#MONTHS_LISTED COLUMN CLEANING
#checking the datatype of the months_listed column
print(f"The data type for months_listed column is : {clean_data['months_listed'].dtype}")
#checking for missing values in the  months_listed column
missing_months_listed = clean_data['months_listed'].isna().sum()
print(f"The number of missing months_listed values is: {missing_months_listed}")
#replacing missing months listed with the mean number of months listed, to one decimal place
clean_data['months_listed'] = clean_data['months_listed'].fillna(round(clean_data['months_listed'].mean(), 1))
#examining changes
missing_months_listed_2 = clean_data['months_listed'].isna().sum()
print(f"The number of missing months_listed values after changes is: {missing_months_listed_2}")

#BEDROOMS COLUMN CLEANING
print(f"The data type for the bedrooms column is : {clean_data['bedrooms'].dtype}")
#checking for missing values in the  months_listed column
missing_bedrooms = clean_data['bedrooms'].isna().sum()
print(f"The number of missing bedrooms values is: {missing_bedrooms}")

#HOUSE TYPE COLUMN CLEANING
print(f"The data type for the house_type column is : {clean_data['house_type'].dtype}")
#checking for missing values in the house_type column
missing_house_type = clean_data['house_type'].isna().sum()
print(f"The number of missing house value is: {missing_house_type}")
#examining the values of house_type
print(clean_data['house_type'].value_counts())
#replacing Det., Semi. & Terr. with Detached, Semi-detached and Terrace
clean_data['house_type'] = clean_data['house_type'] = clean_data['house_type'].replace({
    'Det.': 'Detached',
    'Semi': 'Semi-detached', 
    'Terr.': 'Terraced'
})
#examining changes
print(clean_data['house_type'].value_counts())

#AREA COLUMN CLEANING
print(f"The data type for the area column is : {clean_data['area'].dtype}")
print(clean_data['area'].value_counts())
#removing 'sq.m.' from the area column
clean_data['area'] = clean_data['area'].str.replace('sq.m.', '',regex=False)
#removing any extra whitespace
clean_data['area'] = clean_data['area'].str.strip()
#changing the datatype of area column to float
clean_data['area'] = clean_data['area'].astype('float') 
#checking for missing values in the area column
missing_area = clean_data['area'].isna().sum()
print(f"The number of missing area values is: {missing_area}")
#examing changes
print(clean_data['area'].dtype)
```

# Task 3 

The team at RealAgents have told you that they have always believed that the number of bedrooms is the biggest driver of house price. 

Producing a table showing the difference in the average sale price by number of bedrooms along with the variance to investigate this question for the team.

 - You should start with the data in the file 'house_sales.csv'.

 - Your output should be a data frame named `price_by_rooms`. 

 - It should include the three columns `bedrooms`, `avg_price`, `var_price`. 

 - Your answers should be rounded to 1 decimal place.   

```python
# Use this cell to write your code for Task 3
#creating by_rooms
by_rooms = clean_data[['bedrooms','sale_price']]
#grouping by_rooms by sale_price and calculating the average to create the avg_price and var_price columns
price_by_rooms = by_rooms.groupby('bedrooms')['sale_price'].agg(['mean','var']).reset_index().rename({'mean':'avg_price','var':'var_price'})
#rounding the avg_price to 1 decimal place
price_by_rooms['avg_price'] = round(price_by_rooms['avg_price'],1)
#rounding the avg_price to 1 decimal place
price_by_rooms['var_price'] = round(price_by_rooms['var_price'],1)
#examining the changes
print(price_by_rooms)
```
